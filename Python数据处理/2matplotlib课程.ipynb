## BarÊü±Áä∂Âõæ

```python
n=12 #Êü±Áä∂ÂõæÁöÑÊï∞Èáè
X=np.arange(n)
Y1=(1-X/float(n))*np.random.uniform(0.5,1.0,n) #ÂùáÂåÄÂàÜÂ∏É 0.5-1‰πãÈó¥ÁöÑ12‰∏™ÂùáÂåÄÂàÜÂ∏ÉÊï∞ Â±û‰∫éndarray
Y2=(1-X/float(n))*np.random.uniform(0.5,1.0,n)

plt.bar(X,+Y1,facecolor='#9999ff',edgecolor='white')
plt.bar(X,-Y2,facecolor='#ff9999',edgecolor='white')
#Ê∑ªÂä†ÊñáÂ≠ó
for x,y in zip(X,Y1):
    plt.text(x+0.2,y+0.05,'%.2f'%y,ha='center',va='bottom') # ha=horizontal alignment
for x,y in zip(X,-Y2):
    plt.text(x+0.2,y-0.1,'%.2f'%y,ha='center',va='top') # ha=horizontal alignment
plt.xlim(-.5,n)
plt.xticks(())
plt.ylim(-1.5,1.5)
plt.yticks(())
```

## ContoursÁ≠âÈ´òÁ∫øÂõæ

```python
def f(x,y):
    # ËÆ°ÁÆóÈ´òÂ∫¶ÁöÑÊñπÊ≥ï
    return (1-x/2+x**5+y**3)*np.exp(-x**2-y**2)

n = 256
x = np.linspace(-3,3,n)
y = np.linspace(-3,3,n)

X,Y=np.meshgrid(x,y) # ËΩ¨‰∏∫Á≠âÈ´òÁ∫øÁöÑÁΩëÊ†ºÂùêÊ†á

# Ê∑ªÂä†ÁÉ≠ÊàêÂÉè'Âõæ'f=fill 
plt.contourf(X,Y,f(X,Y),4,alpha=.8,cmap=plt.cm.hot)# 1ÊòØÈ´òÂ∫¶ËßÑÊ®°,Ëá™Âä®Â∞ÜÈ´òÂ∫¶ÂàÜ‰∏∫9Á±ª

#Ê∑ªÂä†Á≠âÈ´ò'Á∫ø'
C=plt.contour(X,Y,f(X,Y),4,colors='black',linewidths=.5)
#Ê∑ªÂä†Ê†áÁ≠æ
plt.clabel(C,inline=True,fontsize=8)

plt.xticks(())
plt.yticks(())
plt
```

## 3DÂõæÂÉè

```python
# Ê∑ªÂä†3DÊ®°Âùó
from mpl_toolkits.mplot3d import Axes3D

fig=plt.figure()
ax=Axes3D(fig)
X = np.arange(-4,4,0.25)
Y = np.arange(-4,4,0.25)
X,Y = np.meshgrid(X,Y)
R = np.sqrt(X**2+Y**2)
Z = np.sin(R)
ax.plot_surface(X,Y,Z,rstride=1,cstride=1,cmap=plt.get_cmap('rainbow')) #rstride=1,cstride=1 Ë°åÂíåÂàóÁöÑË∑®Â∫¶
ax.contourf(X,Y,Z,zdir='z',offset=-2,cmap='rainbow') 
ax.set_zlim(-2,2)
```

# animationÂä®Áîª

```python
from matplotlib import animation
# %matplotlib notebook
fig,ax = plt.subplots()
x = np.arange(0,2*np.pi,0.01)# np.pi=pai
line,=ax.plot(x,np.sin(x))

def aniFunc(i):
    line.set_ydata(np.sin(x+i/100))
    return line,
def init():
    line.set_ydata(np.sin(x))
    return line,

ani = animation.FuncAnimation(fig=fig,func=aniFunc,frames=100,init_func=init,interval=20,blit=False)
```
